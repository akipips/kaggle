### What is Exploratory Data Analysis (EDA)?

- How to ensure you are ready to use machine learning algorithms in a project? 
- How to choose the most suitable algorithms for your data set?
- How to define the feature variables that can potentially be used for machine learning?

**Exploratory Data Analysis (EDA)** helps to answer all these questions, ensuring the best outcomes for the project. It is an approach for summarizing, visualizing, and becoming intimately familiar with the important characteristics of a data set.

### Value of Exploratory Data Analysis

探索的データ分析は、将来の結果が有効であり、正しく解釈され、目的のビジネスコンテキストに適用できるという確実性に近づくことができるため、データサイエンスプロジェクトに役立ちます。このような確実性のレベルは、生データが検証されて異常がないか確認された後にのみ達成され、データセットがエラーなしで収集されたことを確認します。 EDAは、ビジネスの利害関係者やデータサイエンティストにとって明白ではないか調査する価値のない洞察を見つけるのにも役立ちますが、特定のビジネスについて非常に有益な場合があります。


EDAは、機械学習に使用される特徴変数の選択を定義および調整するために実行されます。データサイエンティストがデータセットに慣れると、最初の特徴量が意図した目的を果たさないことが判明する可能性があるため、データサイエンティストはしばしば特徴量エンジニアリングの手順に戻る必要があります。 EDAステージが完了すると、データサイエンティストは、教師ありと教師なしの機械学習に必要な確固とした特徴量セットを取得します。

### Methods of Exploratory Data Analysis
複数の探索的手法を使用して各データセットを探索し、結果を比較することは常に優れています。データセットが完全に理解されると、データサイエンティストがデータ収集とクレンジングフェーズに戻って、目的のビジネス結果に応じてデータセットを変換しなければならない可能性があります。このステップの目標は、データセットを機械学習アルゴリズムで使用する準備ができていることを確信することです。

Exploratory Data Analysis is majorly performed using the following methods:

- 一変量視覚化—生データセットの各フィールドの要約統計を提供します
- 2変量視覚化—データセット内の各変数と対象のターゲット変数との関係を見つけるために実行されます
- 多変量可視化—データセット内の異なるフィールド間の相互作用を理解するために実行されます
- 次元削減—観測間の最大分散を説明するデータのフィールドを理解し、削減されたデータ量の処理を可能にするのに役立ちます。

これらの方法を通じて、データサイエンティストは仮定を検証し、問題の理解とモデルの選択を可能にするパターンを特定し、データが期待どおりに生成されたことを検証します。したがって、各フィールドの値の分布がチェックされ、いくつかの欠損値が定義され、それらを置き換える可能な方法はfです。


探索的データ分析がプロジェクトにもたらすその他の利点
EDAのもう1つの副次的な利点は、データから回答を得ようとしている質問を指定または定義できることです。データサイエンスとAIテクノロジーの活用を始めたばかりの企業は、多くのデータを抱えており、データがビジネスの意思決定にどのような価値をもたらすことができるのかわからないという状況に直面することがよくあります。

ただし、データ分析では常に質問が最初に来ます。ビジネスの利害関係者がデータで解決しようとしている質問がない限り、データが過去のものであるかリアルタイムであるかに関係なく、データ企業がいくら持っているか、利用可能なツールがいくつあるかは関係ありません。 EDAはそのような企業が正しい質問を形式化するのを助けることができます、なぜなら間違った質問であなたは間違った答えを得て、間違った決定をするからです。

#### Why skipping Exploratory Data Analysis is a bad idea?

データサイエンティストは、機械学習の段階への急いで急いで、または単にビジネスの利害関係者に非常に速く印象を与えるために、探索プロセスを完全にスキップするか、非常に浅い作業を行う傾向があります。これは非常に深刻であり、悲しいことに、アマチュアデータサイエンスコンサルティングの「専門家」によくある間違いです。

このような思いがけない振る舞いは、データが歪んで、外れ値や欠損値が多すぎて、プロジェクトに悲しい結果をもたらす可能性があります。:

- generating inaccurate models;
- generating accurate models on the wrong data;
- choosing the wrong variables for the model;
- inefficient use of the resources, including the rebuilding of the model.

一方では、探索的データ分析（EDA）を使用して、質問への回答、ビジネスの前提条件のテスト、さらなる分析のための仮説の生成を行います。一方、モデリング用のデータを準備するためにも使用できます。

これら2つに共通しているのは、必要な答えを得るため、または将来のモデリングの結果を解釈するための直感を開発するために、データをよく理解していることです。

There are a lot of ways to reach these goals as follows:

1. Import the data

2. Get a feel of the data ,describe the data,look at a sample of data like first and last rows

3. Take a deeper look into the data by querying or indexing the data

4. Identify features of interest

5. Recognise the challenges posed by data - missing values, outliers

6. Discover patterns in the data

One of the important things about EDA is  Data profiling. 

**データプロファイリング**は、記述統計によるデータセットの要約に関係しています。データセットをよりよく理解するために、さまざまな測定値を使用したいと考えています。データプロファイリングの目的は、データをしっかりと理解して、後でさまざまな方法でデータのクエリと視覚化を開始できるようにすることです。ただし、これは反復する必要がないという意味ではありません。データプロファイリングはデータセットの要約に関係しているため、データ品質の評価に頻繁に使用されます。データプロファイリングの結果に応じて、データを修正、破棄、または処理する方法を変えることができます。

#### Dendrogram

樹形図を使用すると、変数の完了をより完全に相関させることができ、相関ヒートマップに表示されるペア毎の傾向よりも深い傾向を明らかにします。

```python
msno.dendrogram(train)
```


樹状図は、階層的クラスタリングアルゴリズム（scipyの好意による）を使用して、変数のnullity相関（バイナリ距離で測定）によって変数を互いにビンします。ツリーの各ステップで、残りのクラスターの距離を最小にする組み合わせに基づいて変数が分割されます。変数のセットが単調になるほど、それらの合計距離はゼロに近くなり、それらの平均距離（y軸）はゼロに近くなります。

このグラフを解釈するには、上から見た視点からグラフを読んでください。ゼロの距離で一緒にリンクされたクラスターリーフは、互いの存在を完全に予測します。1つの変数は、別の変数が満たされると常に空になる場合があります。この特定の例では、樹状図は、必要であり、したがってすべてのレコードに存在する変数を接着します。

クラスターリーフはゼロに近く分割されますが、ゼロではありませんが、お互いを非常によく予測しますが、それでも不完全です。データセットを独自に解釈して、これらの列が実際にnullで一致する、または一致するはずである場合、クラスターリーフの高さによって、絶対的に、レコードが「不一致」または誤ってファイルされる頻度がわかります。傾向がある場合、入力またはドロップする必要がある値の数です。

マトリックスと同様に、この構成では最大50のラベル付き列のみが快適に表示されます。ただし、樹状図は、水平構成に切り替えるだけで非常に大きなデータセットをよりエレガントに処理します。
